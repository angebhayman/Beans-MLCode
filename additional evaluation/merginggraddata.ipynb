## Model

```python
df_cleaned
```

```python
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestRegressor
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder
from sklearn.impute import SimpleImputer
from sklearn.metrics import mean_squared_error, r2_score

# Assuming 'average_rating' is the target variable, and other columns are features
# Separate features (X) and target (y)
X = df_cleaned.drop(columns=['average_rating'])
y = df_cleaned['average_rating']

# Define categorical and numeric columns
categorical_cols = ['DBN', 'School Name']  # Replace with actual categorical columns
numeric_cols = X.select_dtypes(include=['int64', 'float64']).columns.tolist()  # All numeric columns

# Define the preprocessing for both numeric and categorical data
# For numeric data, we will fill missing values with the median
# For categorical data, we will use one-hot encoding after filling missing values
preprocessor = ColumnTransformer(
    transformers=[
        ('num', Pipeline([
            ('imputer', SimpleImputer(strategy='median'))
        ]), numeric_cols),
        ('cat', Pipeline([
            ('imputer', SimpleImputer(strategy='constant', fill_value='missing')),
            ('onehot', OneHotEncoder(handle_unknown='ignore', drop='first'))
        ]), categorical_cols)
    ])

# Create a pipeline that first transforms the data and then applies the model
model = Pipeline(steps=[
    ('preprocessor', preprocessor),
    ('regressor', RandomForestRegressor(n_estimators=100, random_state=42))
])

# Split the data into train and test sets
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)

# Train the model
model.fit(X_train, y_train)

# Make predictions
y_pred = model.predict(X_test)

# Evaluate the model
mse = mean_squared_error(y_test, y_pred)
r2 = r2_score(y_test, y_pred)

print(f'Mean Squared Error: {mse:.2f}')
print(f'R-squared: {r2:.2f}')
```
